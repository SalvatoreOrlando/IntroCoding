#### Ex 1
Write a function that takes a date in the `“dd-MMM-yy”` format (a string) and returns a tuple of integers  `(y, m, d)`,   where a tuple in an immutable list). 
*HINT: Create a dictionary suitable for decoding month names to numbers. Use the string operations to split the date into 3 items using the `'-'` character as a separator. Correct the year to include all of the digits (the years go from 1950 till 2010).*





```python
month = {'Jan':1, 'Feb':2, 'Mar':3, 'Apr':4, 'May':5, 'Jun':6, 'Jul':7, 'Aug':8, 'Sep':9, 'Oct':10, 'Nov':11, 'Dec':12}

def transform(date):
    li = date.split('-')
    day = int(li[0])
    mon = month[li[1]]
    year = int(li[2])
    if year <= 10:
        year = 2000 + year
    else:
        year = 1900 + year
    return (year, mon, day)
    
date1 = "30-Apr-09"
date2 = "14-Jun-70"

print(transform(date1))
print(transform(date2))
```

Write a function that reads the file and stores the pairs in a dictionary. Write another function that reads a name, and lookup the dictionary for the input name, and returns either the associated phone number or -1. 

```python
def read_dict(filename, dict):
    f = open(filename, 'r')
    while True:
        line = f.readline()
        if (len(line) == 0):
            f.close()
            return
        li = line.split()
        dict[li[0]] = li[1]
    f.close()

def lookup(name, dict):
    if name in dict:
        return dict[name]
    else:
        return -1

'''
n = int(input("how many pairs?: "))
f = open('addr_book.txt', 'w')
for i in range(n):
    name = input("Name: ")
    phone = input("Phone num.: ")
    f.write(name + ' ' + phone + '\n')
f.close()
'''
dict = {}
read_dict('addr_book.txt', dict)
print('Sal: ', lookup('Sal', dict))
print('Ant: ', lookup('Ant', dict))
print('Luca: ', lookup('Luca', dict))

    
```
